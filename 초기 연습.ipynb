```python
import numpy as np

#간단한 벡터(양자 상태 |0>)
state_0 = np.array([1, 0])
print("Quantum state |0>:", state_0)

# 간단한 행렬 (Pauli-x 게이트)
X_gate = np.array([[0, 1], [1,0]])
print("\nPauli-X gate:")
print(X_gate)

# 행렬-벡터 곱셉
result = X_gate @ state_0
print("\nX|0> =", result, " <- 이게 |1> 상태!")
```

```python
import matplotlib.pyplot as plt
import numpy as np

#간단한 파동 함수 그리기
x = np.linspace(0, 2*np.pi, 100)
psi = np.sin(x)

plt.figure(figsize=(10,4))
plt.plot(x, psi, 'b-', linewidth=2)
plt.title('Wave Function (파동함수 예시)', fontsize=14)
plt.xlabel('x')
plt.ylabel('ψ(x)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 1. 리스트와 반복문
numbers = [1,2,3,4,5]
squares = [n**2 for n in numbers]
print("제곱:", squares)

# 2. 함수 만들기
def quantum_state(alpha, beta):
    """정규화된 양자 상태 생성"""
    norm = np.sqrt(abs(alpha)**2 + abs(beta)**2)
    return np.array([alpha/norm, beta/norm])

# 테스트
state = quantum_state(1, 1) #균등 중첩 상태
print("Superposition state:", state)
print("확률:", np.abs(state)**2)

# 3. Numpy 배열 연산
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print("행렬 곱:", A@B)
print("고유값:", np.linalg.eigvals(A))
```

```python
#Y 게이트와 Z 게이트도 만들어보기
Y_gate = np.array([[0, -1j],[1j, 0]])
Z_gate = np.array([[1,0],[0,-1]])

#중첩 상태 만들기
superposition = np.array([1,1]) / np.sqrt(2) # (|0> + |1>)/√2
print("중첩 상태:", superposition)

#각 게이트 적용해보기
print("\nX * 중첩 =", X_gate @ superposition)
print("\nY * 중첩 =", Y_gate @ superposition)
print("\nZ * 중첩 =", Z_gate @ superposition)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1차원 무한 우물 속 파동함수
x= np.linspace(0,1,200)
n_values = [1,2,3]

plt.figure(figsize=(12,4))

for i, n in enumerate(n_values, 1):
    plt.subplot(1,3,i)
    psi = np.sqrt(2)*np.sin(n*np.pi*x)
    prob = psi**2 #확률 밀도

    plt.plot(x, psi, 'b-', label='ψ(x)', linewidth=2)
    plt.fill_between(x, 0, prob, alpha=0.3, color='red', label='|ψ(x)|²')
    plt.title(f'n = {n}', fontsize=14)
    plt.xlabel('x')
    plt.ylabel('ψ')
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
#Bloch 구 좌표계에서 상태 벡터 (2D 단순화 버전)
import matplotlib.pyplot as plt
import numpy as np

theth = np.linspace(0, 2*np.pi, 100)

# |0>, |1>, 중첩 상태 표현
states = {
    '|0>':[0,1],
    '|1>':[0,-1],
    '|+> = (|0>+|1>)/√2':[1,0],
    '|-> = (|0>-|1>)/√2':[-1,0]}

plt.figure(figsize=(8,8))
circle = plt.Circle((0,0), 1, fill=False, color='gray', linestyle='--')
plt.gca().add_patch(circle)

for label, [x,y] in states.items():
    plt.arrow(0,0,x,y, head_width=0.1, head_length=0.1, fc='blue', ec='blue', linewidth=2)
    plt.text(x*1.2, y*1.2, label, fontsize=12, ha='center')

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5,1.5)
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.title('양자 상태 표현 (단순화)', fontsize=16)
plt.xlabel('x')
plt.ylabel('Z')
plt.axis('equal')
plt.show()
```

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

#한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 1. 양자회로 만들기 (큐비드 1개)
qc = QuantumCircuit(1, 1) # (양자비트 1개, 고전비트 1개)

# 2. X게이트 추가(비트 뒤집기)
qc.x(0)

# 3. 측정
qc.measure(0,0)

# 4. 회로 그리기
print(" 회로 다이어 그램:")
print(qc.draw())

# 5. 시뮬레이터에서 실행
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=1000)
result = job.result()
counts = result.get_counts()

# 6. 결과 시각화
print(f"\n 측정결과: {counts}")
plot_histogram(counts, title = '측정 결과')
plt.show()
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# 1) 상수 함수 (f(0)=f(1)=0) -아무것도 안 함
def constant_oracle(qc):
    pass # 아무 게이트도 안 넣음

# 2) 균형 함수 (f(0)≠f(1)) - CNOT으로 구현
def balanced_oracle(qc):
    qc.cx(0,1)

# Deutsch 알고리즘
def deutsch_algorithm(oracle):
    qc= QuantumCircuit(2,1)

    #초기화
    qc.x(1) #보조 큐비트를 |1>로
    qc.h(0) #입력 큐비트 중첩
    qc.h(1) #보조 큐비트도 H

    #오라클 적용(블랙박스)
    oracle(qc)

    #간섭 단계
    qc.h(0)

    #측정
    qc.measure(0,0)

    return qc

# 실험 1: 상수 함수
qc1 = deutsch_algorithm(constant_oracle)
result1 = AerSimulator().run(qc1, shots=100).result()
print("상수 함수:", result1.get_counts())

# 실험 2: 균형 함수
qc2 = deutsch_algorithm(balanced_oracle)
result2 = AerSimulator().run(qc2, shots=100).result()
print("균형 함수:", result2.get_counts())
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

def bernstein_vazirani(secert_string):
    n = len(secert_string)
    qc = QuantumCircuit(n+1, n)

    # 1) 초기화
    qc.x(n) # 보조 큐비트
    qc.h(range(n+1)) # 전부 중첩

    # 2) 오라클 (비밀 문자열 인코딩)
    for i, bit in enumerate(reversed(secert_string)):
        if bit == '1':
            qc.cx(i, n)

    # 3) 간섭
    qc.h(range(n))

    # 4) 측정
    qc.measure(range(n), range(n))

    return qc

# 실험 : 비밀 문자열 "101"
secert = "101"
qc = bernstein_vazirani(secert)
result = AerSimulator().run(qc, shots=1).result() # 딱 1번만!
print(f"비밀 문자열: {secert}")
print(f"측정결과: {result.get_counts()}")
print(qc.draw())
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import math

def grover_search(target='11'):
    """target: 찾으려는 항목 (예: '1')"""
    qc = QuantumCircuit(2,2)

    # 1) 초기화 : 균등 중첩
    qc.h([0,1])

    # 2) Grover 반복 (√N번 - 여기서는 1번이면 충분)
    # Oracle : target을 마킹
    if target == '11' :
        qc.cz(0,1) # |11>에 마이너스 부호
    elif target == '10':
        qc.x(1)
        qc.cz(0,1)
        qc.x(1)
    # ... (다른 경우도 비슷)

    # Diffusion : 확률 증폭
    qc.h([0,1])
    qc.z([0,1])
    qc.cz(0,1)
    qc.h([0,1])

    # 3) 측정
    qc.measure([0,1], [0,1])

    return qc

# 실험: '11'을 찾아보자
qc = grover_search(target='11')
result = AerSimulator().run(qc, shots=100).result()
print("검색 대상: 11")
print("측정 결과:", result.get_counts())
print(qc.draw())
```
